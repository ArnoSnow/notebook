```python
# add.reduce，虽然我理解和add在部分情况下值是相等的，但是实用层面上我知道了行总计，列总计的用法
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])

result = np.add.reduce(arr, axis=0) # 等于0相当于行总计
print(result)  # 输出: [5 7 9]
result = np.add.reduce(arr, axis=1) # 等于1相当于列总计
print(result)  # 输出: [ 6 15]
```

### 三、统计类函数与数据维度
###### 1、统计类函数
最常见的一种统计方法，描述性统计，因为不能描述上千条数据，只能进行聚集函数，来统一描述一批数据的情况

```python
# 上面学习的是对于数组与整数/其他数组之间的计算方式，现在是对于一个数组里面包含的数据执行统计算法
import numpy as np
score = np.array([70,80,96])
print( np.sum(score) )
print( np.average(score) ) # weights参数可以指定其他数组/列表为权重数组，计算加权平均值，仅有函数用法，无数组方法
print( np.mean(score) ) # 仅计算算数平均值
print( np.std(score) ) # 标准差，有助于知道每个数据和整个数据集的平均值的离散程度或变异程度，从而判断正态分布
print( np.var(score) ) # 样本方差
print( np.min(score) ) # 最小值
print( np.max(score) ) # 最大值
print( np.ptp(score) ) # 极差，算最大值与最小值之间的差值，即峰-峰值。
# print( np.percentile(score) ) # 计算百分位数
print( np.median(score) ) # 寻找中位数
```

```python
# 详细解释下上面的标准差，样本方差等概念
import numpy as np
# 为了避免无法理解部分知识，与此补充数学相关基础知识
# 计算式如下：s =√[Σ(xi - x̄)²/ (n - 1)]
# 步骤一 x̄ 表示样本平均值 ，算法：x̄ = (x₁ + x₂ + x₃ + ... + xₙ) / n
arr1 = np.array([1,2,3,4,5])
import statistics as st
avg = st.mean(arr1)

# 步骤二 每个元素都和平均值做减法完成xi - x̄，算下每个元素和平均值的差值
arr1 = arr1 - avg # 

# 步骤三 每个差值做平方下
arr1 = arr1**2

# 步骤四 求和(xi - x̄)² Σ是希腊字母大写的西格玛（Sigma），在数学中用作求和符号
arr1.sum()

# 步骤五 样本标准差和整体标准差，存在一个自由度的说法，
# 当你的样本较少时，实际也不可能获取所有样本，所以为了近似整体，样本标准差会选择除以n-1，整体却除以n
# 标准
print( np.sqrt( arr1.sum()/(len(arr1)-1) ) )
# 整体
print( np.sqrt( arr1.sum()/len(arr1) ) )


print(np.std([1,2,3,4,5])) # # 这个结果和直接调用np.std是一致的,说明算的是整体，也就是标准差
print(np.var([1,2,3,4,5])) # # 这个结果和print( np.sqrt( arr1.sum()/(len(arr1)-1) ) ) 的平方数值一样，所以还不是样本标准差的关系
```

### 三、统计类函数与数据维度
###### 2、数据维度
维度的本质是特征或者属性，多一个特征多一个维度
可以理解为sql的分组group by，分成期中考试聚合每个用户来看，或者某个用户的多个科目的平均值，聚合了科目，但是分组和维度又有差别，因为最终落库还是表格类型，最多就是二维了
###### 3、维度在数组的逻辑下遵循以下原则
（1）同一个数据集里的所有内容最终一定涵盖在一对方括号里面
（2）从最左边或者最右边去数一共有几层方括号就代表有几个维度；从左到右，从外到内依次为0号维度，1号维度，2号维度，3号维度，见下图案例
（3）axis参数指定维度，指定的维度就消灭它也即通过聚合的方式降低维度，比如消灭0号维度，我们就无法知道学期每个学期的情况，但可以知道科目、姓名、成绩类型等其他信息，所以聚合时要写axis = 0 ，二维表格下，这个非常好理解，就是行汇总和列汇总；问题在于高维情况

### 四、numpy的广播规则 broadcast
前言
前文中通过整体运算和相同形状，对位计算这两个原则，已知同形数组之间已经可以计算，一个数组和一个数字也可以计算\n
广播规则解决的就是不同形状的数组能不能计算的问题，这个问题的答案和上面学习过的维度又有关系
###### 1、广播规则的本质可以概括为依据规则脑补不同形的数组为同形，在按照同形数组对位计算的原则计算
#脑补规则：同维需同长，除非长唯一；此处的同维是指从内向往，从有向左编号的同一个编号维度必须同长，如果不同长，为了触发原则，必须长唯一，如果比对方多维度反而不要紧

### 六、布尔掩码，关系运算与布尔值的灵活使用 关系运算比较特殊，包含选择，投影，链接，交集，并集，差集
###### 2、掩码快速抽取符合规则的数组元素

```python
# 什么是掩码？一个尺子符合条件的放出来，不符合条件的遮住，如何知道一个是否符合，前面的比较运算已经揭露了
import numpy as np
arr1 = np.random.randint(100,150,(3,5))
print(arr1)
print(arr1 > 120) # 可以广播
print(arr1[arr1>120]) # 掩码，看起来就是切片了，条件切片了
# 既然可以一个条件切片，当然可以多个条件并列切片，与 & 或 \ 非 ~
arr2 = np.array([12,10,8,6,5,3,2,1])
arr2[(arr2>10) | (arr2<3)] # 每个条件用括号括起来，以免发生优先级错误
arr2[~(arr2>10)] # 非
# 为什么使用常规的and or not会导致错误？必须使用符号？因为python自带的强制转化逻辑值，但是array的数组在元素超过1个时强制不给转，python就算不出来就报错
# 这些特殊符号被称之为位运算符，基于算数运算+广播原则对于两个数组里面的值对位计算并返回一个新数组，
# True + True = True True+ False = False  False+ False = False 从而实现同时满足多个，或者其中1个，非等操作

# 位运算符的运算优先级别高于关系运算符，优先计算位运算，其次才是关系的>，<，=这些符号，所以必须加括号，不加就会乱套
```

```python
# 增加点问题，掩码一般用于同一个数组，自己的元素是否判定后，切片出是true的值
# 如果形状不一样的，会通过广播规则来适配吗
arr1 = np.array([[1,2,3,4],[5,4,3,2]])
arr2 = np.array([1,2])
print(arr1[arr2>1]) # 会报错
print(arr2>1)
# 广播原则要求从内到外，同维需同长，除非长唯一，但是这个就不满足，但是一样计算了，
# 他的做法是去除了false 的第一行，满足的是行是两行，给出了false一个值，true 一个值，同样完成计算，不符合广播但是类似广播
```

###### 1、花式索引的基础用法

```python
import numpy as np
# 步骤1 准备个演示数组
list1 = [[300,150,200,200],[300,150,250,300],[200,80,100,100]]
a = np.array(list1)
print(a,end = '\n\n') # 假设行为商品种类，列为月份销量

# 步骤2 花式索引展示
print(   a[[1,2],[0,1]]   ,end = '\n\n') # (1,0)组成一个元素坐标，（2,1）组成一个元素坐标，根据坐标即可生成

# 步骤3 花式做引同样适用广播原则
print(   a[ [1,2],2 ]   ,end = '\n\n')  # 此处自动扩张2为[2,2]从而构成坐标（1,2）（2,2）
```

### 八、花式索引 Fancy Indexing
###### 3、烧脑同时的例子

```python
# 步骤1 准备个演示数组
list1 = [[300,150,200,200],[300,150,250,300],[200,80,100,100]]
a = np.array(list1)
print(a,end = '\n\n') # 假设行为商品种类，列为月份销量

# 步骤2 计算每个月份所有产品的销售量总和
s =  a.sum(axis = 0) 
print( '每个月的销售量总和是：', s ,end = '\n\n')

# 步骤3 对于新生的s数组去取里面大于600的列值
列号 = np.where(s>=600)[0] #注意返回的是元祖
print(列号,end = '\n\n')

# 步骤4 对于这些表现良好（销售量总和大于等于600的月份）每个产品的销量捞出来看看明细
print(  a[  : ,列号       ]    ,end = '\n\n'  )# 这是看所有产品
print(  a[  [[0],[2]],列号   ] ,end = '\n\n'  )# 这是对于代表是什么产品的行也做切片，切出国定的某x个产品
```

```python
# 课后习题1
# 生成一个范围在100~200，共4行6列的数组，获取 第2行3列、第1行5列、第4行6列 的元素。
import numpy as np
arr1 = np.random.randint(100,201,(4,6))
print(arr1,end='\n\n')

# 开始基础的Fancy Indexing
print(arr1[1,2],arr1[0,4],arr1[3,5])
print(arr1[[1,0,3],[2,4,5]])
```

```python
# 课后习题3
# 生成一个由逻辑值False和True为元素的5行4列的数组：
import numpy as np
arr1 = np.where(np.random.randint(0,2,(5,4))==0,False,True)
print(arr1,end='\n\n')
#1. 获取第 2~4 行中位于第 1、3 列的数据。
print(arr1[ [ [1],[2],[3]  ],[0,2]    ],end='\n\n')
print(arr1[ 1:4,[0,2]    ],end='\n\n') # 混合使用，可以不必写的特别复杂的广播处理逻辑了
#2. 获取前3列中位于第 3、5 行的数据。 
print(arr1[ [ [2],[4]  ],:3],end='\n\n')
```

```python
# 课后习题5
# 某地区部分行业 2007年~2018年 的产业增长值数据，各行为 2007~2018 ，各列为：运输业,零售业,餐饮业,金融业,房地产。
import numpy as np
arr1 = np.loadtxt(r"D:\1-script\3-PYTHON\data存储jupyter用的个文件类型数据\python_03_09.txt",delimiter = ',')
print(arr1,end='\n\n')
# 1. 获取各行业增长值总和大于1000万的增长值数据；
列号 = np.where(arr1.sum(axis = 0)>1000*10000)[0]
print(arr1[:,列号],end='\n\n')
# 2. 获取各年年均增长值小于200万的增长值数据。
行号 = np.where(arr1.mean(axis = 1)<200*10000)[0]
print(arr1[行号,:],end='\n\n')
```
