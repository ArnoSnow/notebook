```python
# 2.1.2 私有属性读（getter）写(setter)方法一 显性并非属性而是方法，指写的时候带()
class MiShu1():
    def __init__(self):
        self.name = '无名'       # 公有属性
        self.__location = '上海' # 私有属性，在底层的名字会被python修改为其他东西，但在类别内还是可以用这个名称调用
        
    # getter方法读之
    def get_loc(self):
        return self.__location
    
    # setter方法改之，设置私有属性却允许修改可以用在用户使用系统时，允许自定义自己的属地，但是有一定的限制给出
    def set_loc(self,loc):
        if loc not in ['上海','北京']:
            print('属地异常，请重新输入')
        else:
            self.__location = loc

            
m1= MiShu1()
m1.set_loc('北京')
m1.get_loc() # 北京

m1.set_loc('东京')
m1.get_loc() # 北京
```

```python
# 2.1.3 私有属性的读写方法二(@标注法 装饰器原理实现) 隐性 和普通属性看起来一致的用法

class MiShu1():
    def __init__(self):
        self.name = '无名'       # 公有属性
        self.__location = '上海' # 私有属性，在底层的名字会被python修改为其他东西，但在类别内还是可以用这个名称调用
    
    @property # 这个属于函数的装饰器知识，我们知道方法就是函数，这个装饰器的含义是将类中的方法变成属性调用。
    def location (self): # 注意为了伪装这里必须取去除__下滑线的属性名称
        """对外看起来像是公共属性的getter方法"""
        return self.__location
    
    @location.setter # 上面建立的私有属性伪装名，也是这里.setter之前的名称
    def location(self, value):  # 为了伪装这里必须取去除__下滑线的属性名称
        """对外看起来像是公共属性的setter方法"""
        # 这里可以添加一些设置私有属性前的检查逻辑
        self.__location = value

m1= MiShu1()
print(m1.location)
m1.location = '北京'
print(m1.location)
```

```python
# 4.定义类实例的 展示问题
# 4.1 什么是展示
# (1) 展示的本质是 print()/str()/repr()  这些函数时，调用类定义过的属性__str__/__repr__去描述这个对象是啥 
# (2) 面向开发者，调用repr() ，python有效的表达式，可以再次生成一个同样的对象，返回__repr__
# (3) 面向使用者，调用print()/str()，非有效的表达式，只为了便于理解存在，优先返回__str__，无就返回__repr_，这是必定有的，即使你不定义
```

```python
# 4.2 展示不写的情况下，python的解释
# (1) __main__代表的是主模块定义的类，非导入的
# (2) Human object 代表的是Human类的对象
# (3) at 0x0000024B361B6E00>，代表的是内存地址

# 定义一个类别，但是不定义__str__ 和 __repr__
class Human():
    def __init__(self,name = '无名',age=18):
        self.name = name
        self.age = age
    def self_introduction(self):
        print(f'my name is{self.name},my age is {self.age}')
        
# 生成实例对象
human1 = Human()

# 测试各种展示
print(human1) # <__main__.Human object at 0x0000024B361B6E00>
print(str(human1)) # <__main__.Human object at 0x0000024B366560E0>
print(repr(human1)) # <__main__.Human object at 0x0000024B366560E0>
print(hex(id(human1))) # 0x24b377dfdc0 节省内存地址减少了前导零
```

```python
# 4.3 定义__str__ 和 __repr__
class Human():
    def __init__(self,name = '无名',age=18):
        self.name = name
        self.age = age
    # 增加定义str
    def __str__(self):
        return f'代表一个人类，名字是{self.name}，年龄是{self.age}，通常会做自我介绍'
    # 增加定义repr
    def __repr__(self):
        return f'Human(name = {self.name},age = {self.age})'
    def self_introduction(self):
        return(f'my name is{self.name},my age is {self.age}')

# 实例化
human1 = Human()

# 感受下改变后的
print(str(human1))
print(repr(human1))
```

### 自定义类Class
###### 二、使用定义的类
（1）实例化对象 （2）定义语句外动态添加改写属性 （3）定义语句外动态添加改写方法    
总结：    
1. 类属性被所有实例共享，修改后所有实例均受影响。    
2. 实例属性是每个实例独有的，修改构造方法 (__init__) 不会影响已经创建的实例的属性，只会影响之后创建的实例的初始化过程。    
3. 方法是在运行时解析的，修改类的方法定义会影响所有实例上该方法的行为，无论这些实例是在修改之前还是之后创建的。   
4. 以上的本质就是，python优先读取实例的属性/对象，没有才读取类的，读取类的情况下，如果改了类就会带动实例改
5. 部分python自由类别无法被动态添加，因为底层是c语言

```python
# 1.类的实例化
# 语法：变量名m1 = 类名() MiShu()  m1即为类的一个实例
# m1 = MiShu()
```

```python
# 2.动态添加属性
# 语法：对象.属性 = xxxx 属性是啥，你命名的变量，只是只想一个非函数所在的地址

# 2.1 动态添加实例属性

# 定义类别
class MiShu():
    def __init__(self):
        self.name = '无名'
# 实例化
m1 = MiShu()
# m1.age = 18 

# 动态添加属性的本质是 类别/实例.变量 = 值
# print(m1.age)  # 18

# 2.2 这个是给类动态加属性，所有的子类都是一个地址，但是不会改m1的，因为m1是重新赋值的
MiShu.age = 25
m2 =  MiShu()
print(m2.age)  # 25
print(m1.age) # 修改类属性，即使是早于动态添加之前生成的实例也会生效哦，因为每次获取属性都必定读取一次自己的类
```

```python
# 3.动态添加方法
# 语法: 对象.方法 = 函数 方法是啥，仍旧是一个你命名的变量，只是指向一个函数所在的地址

# 3.1 动态添加实例的方法

# 这个是使得方法从属于某个对象的方式，隶属于type库
from types import MethodType

# 定义类别
class MiShu:
    def __init__(self):
        self.name = '无名'
        
# 定义函数作为参数预备
def attack(self):
    return f'{self.name}发起了攻击'

# 实例化
m1 = MiShu()

# 将 attack 函数绑定为 m1 实例的方法
m1.attack = MethodType(attack, m1) # 传递函数作为参数，传递对象作为要从属的对象,如果直接写m1.attack = attack，就需要手动传递self参数

# 现在可以像调用正常实例方法那样调用 attack 了
print(m1.attack())  # 输出: 无名发起了攻击

# 对对象实例加入的属性不会影响类，更不会影响类制造的其他实例
m2 =  MiShu()
# m2.attack # AttributeError
```

```python
# 3.2 动态添加类的方法
# 定义类别
class MiShu:
    def __init__(self):
        self.name = '无名'
        
# 定义函数作为参数预备
def attack(self):
    return f'{self.name}发起了攻击'

# 生成实例，并实现3.1的操作，实例增加方法
m1 = MiShu()
from types import MethodType
m1.attack = MethodType(lambda self:f'{self.name}发起了自杀攻击', m1)
print('实例拓展了方法：',m1.attack())

# 函数赋值给变量 ，这就是动态添加类的本质，类/对象.变量 = 函数
MiShu.attack = attack
print('早于类覆盖先创建的实例方法是否变更',m1.attack()) # '无名发起了自杀攻击'  不会变更，因为优先读取实例的方法

m2 = MiShu()
print('类覆盖方法后创建的',m2.attack())  # 无名发起了攻击 发生了改变

# 同理，构造方法__init__也可以覆盖，但是之前创建的不会变了， 因为构造方法相当于为实例添加了属性 优先读取实例的
def new__init__(self):
    self.name = '有好多名'

# 构造方法，指定属性的__init__,也可以覆写，但是m1不受影响，m2才会生效
MiShu.__init__ = new__init__
m2 = MiShu()
print('类覆盖方法后创建的',m2.name)
print('类覆盖方法后前创建的且没有覆盖属性但也不会改变',m1.name)
```

```python
# 1. 继承演示 clas B(A): 就会继承A类
# 定义对象老师
class Teacher():
    def __init__(self,name = '无名',age = 28):
        self.name = name
        self.age = age
    def settle_problem(self):
        return '这个问题我给他解决了，阔呀'

# 定义对象学生，继承老师的    
class Student(Teacher):
    pass

# 实例化对象
t1 = Teacher()
s1 = Student()

# 老师有的能力学生也有了，这就是继承
s1.settle_problem()
```

```python
# 2.继承的属性与方法覆盖/扩展
# 2.1 继承是可以覆盖的 方式就是在写一次同名方法 ；继承属性是可以扩展的/也可以覆盖的，方法是写同名发放，并写继承父类语句
class Teacher():
    def __init__(self,name = '无名',age = 28):
        self.name = name
        self.age = age
        
    def settle_problem(self):
        return '这个问题我给他解决了，阔呀'

# 定义对象学生，继承老师的    
class Student(Teacher):
   
    def __init__(self,name,age,gender = '男'): # 使用构造法动态添加属性，除非在里面在写一句父类的继承法，否则会整个覆盖父类的属性
        self.gender = gender # 这个是子类特有属性
        #super().__init__(name,age) # 这个是子类继承父类的属性写法一
        Teacher.__init__(self,name,age) # # 这个是子类继承父类的属性写法二
        # 本质只是继承了如下的两句话，参数及其预设值通通需要重定义，并且需要你分配下属性给super() super的用途是自动获得继承者的类名
#         self.name = name
#         self.age = age
            
    def settle_problem(self):
        return '学不会，这就run'
    pass

# 实例化对象
t1 = Teacher()
s1 = Student('有名','18')

# 覆盖方法/属性
# s1.age() # 'Student' object has no attribute 'age' 报错了，因为覆盖
s1.settle_problem() # 学不会，这就run 因为覆盖了，所以不是原来那句
t1.settle_problem() # 这个问题我给他解决了，阔呀  老师的还是不会改变

s1.age
```

```python
# 2.2 私有属性能否继承？继承了，但无法正常调用，所以如果希望子类去访问但不希望外部去访问就延伸了 保护属性 
# 私有属性能够继承，但是继承的是在Teacher类中的私有，名称就会是_Teacher__name；但子类使用时默认组合的私有属性名称是_Student__name，但这个是没有的，就会出错
# 定义老师
class Teacher():
    def __init__(self,name = '无名',age = 28):
        self.__name = name # 私有属性
        self.age = age
        
# 定义继承老师的学生
class Student(Teacher):
    def __init__(self,name = '有名',age=12,gender= 'man'):
        super().__init__(name,age) # 继承父类属性，也就意味着私有属性也在里面应该被继承了？
        self.gender = gender       # 扩展的属性

# 实例化
t1 = Teacher(name = '张三',age= 32) 
s1 = Student()

# 尝试读取私有属性
print('违规获取私有属性：',t1._Teacher__name) # 张三是能拿到的
# print('违规获取私有属性：',s1._Student__name) # AttributeError 为什么？
print('违规获取私有属性：',s1._Teacher__name) # 有名可以拿到，并且确实不受父类的可选参数默认值影响
```

### 自定义类Class
###### 五、对象的协调调用 消息机制
mud 古早文字交互游戏作为示例，例子中我特意使用了私有属性，为了测试调用私有属性的方式，结论如下          
（1）私有属性的调用，会极大程度的影响类的继承，导致很多方法容易出错            
（2）为了保证松耦合，减少类与类通信的问题，即使在类的内部访问私有属性的时候最好也使用公共接口保证数据自动校验一致性       

```python
from random import choice,sample
# 定义 Role类
class Role():
    def __init__(self,name,blood,power):        
        self.__name =  name
        self.__blood = blood
        self.__power = power
    
    def __dead(self):
        if self.get_blood()<=0:
            print(f'{self.get_name()}已经死亡')
            return True
        else:
            # print(f'{self.get_name()}还能战斗')
            return False
    
    # 即使同一个类别的实例，不可以调用对方的私有属性，所以只能定义get和set来解决，这显得标注法更有价值
    def attack(self,enemy):
        print(f'{self.get_name()}发起了攻击，攻击力是{self.get_power()}，{enemy.get_name()}被命中，原血量是{enemy.get_blood()}')
        enemy.set_blood (enemy.get_blood() - self.get_power()) 
        print(f'{enemy.get_name()}剩余血量是{enemy.get_blood()}')
    
    # 公共属性准备
    def get_name(self):
        return self.__name
    def get_blood(self):
        return self.__blood
    def get_power(self):
        return self.__power
    
    # 公共方法准备
    def set_name(self,name):
        if isinstance(name,str):
            self.__name = name
        else:
            print( '名字不规范')
            
    # 私有属性不可以被其他类修改，如果你的属性被别的实例调用，必须得有公共方法
    def set_blood(self,blood):
        if isinstance(blood,int):
            self.__blood = blood

    def dead(self):
        return self.__dead()


# 定义玩家类 继承的类私有属性/方法过多 等于假继承了捏
class Player(Role):
    
    # 私有属性，继承不了，重写
    def __init__(self,name,blood,power):
        # super().__init__(name,blood,power) # 不可以这样写哦，父类私有属性继承会造成类名错误找不到的
        self.__name =  name
        self.__blood = blood
        self.__power = power
        
    # 私有方法，继承不了，重写
    def __dead(self):
        if self.get_blood()<=0:
            print(f'{self.get_name()}已经死亡')
            return True
        else:
            # print(f'{self.get_name()}还能战斗')
            return False
        
    # 本身也需要重新定义，直接覆盖
    def attack(self,enemy):
        # 控制输入是否合法
        user_input = None
        
        # 不合法就一直运行
        while user_input not in ['1', '2']:
                user_input = input('请选择攻击方式：输入【1】，疯狗攻击，输入【2】,普通攻击')
                if user_input == '1':
                    # 确定力量随机数
                    power = self.get_power() * choice(range(1,5))
                    print(f'{self.get_name()}发起了疯狗攻击，攻击力是{power}，{enemy.get_name()}被命中，原血量是{enemy.get_blood()}')
                    enemy.set_blood(enemy.get_blood() - power)
                    print(f'{enemy.get_name()}剩余血量是{enemy.get_blood()}')
                    
                    # 确定扣血随机数
                    hurt = int(self.get_power() * choice(range(1,5)) * 0.3 ) # 下面要求判断整数，不写else发现不是整数也不会报错也不会扣血
                    print(f'{self.get_name()}被疯狗攻击反噬，反噬力是{hurt}，原血量是{self.get_blood()}')
                    self.set_blood (self.get_blood()-hurt)
                    print(f'{self.get_name()}剩余血量是{self.get_blood()}')
                    
                elif user_input == '2':
                    print(f'{self.get_name()}发起了攻击，攻击力是{self.get_power()}，{enemy.get_name()}被命中，原血量是{enemy.get_blood()}')
                    enemy.set_blood(enemy.get_blood() - self.get_power())
                    print(f'{enemy.get_name()}剩余血量是{enemy.get_blood()}')   
     
    # 公共属性准备
    def get_name(self):
        return self.__name
    def get_blood(self):
        return self.__blood
    def get_power(self):
        return self.__power
    
    # 公共方法准备
    def set_name(self,name):
        if isinstance(name,str):
            self.__name = name
        else:
            print( '名字不规范')
            
    # 私有属性不可以被其他类修改，如果你的属性被别的实例调用，必须得有公共方法
    def set_blood(self,blood):
        if isinstance(blood,int):
            self.__blood = blood

    def dead(self):
        return self.__dead()
        
# 实例 role、player
王五 = Role(name = '王五',blood = 100, power = 60)
赵三 = Role(name = '赵三',blood = 10, power = 1000)
李六 = Role(name = '李六',blood = 200, power = 30)
金蛇郎君 =  Player(name = '金蛇郎君',blood = 60, power = 50)

# 调用方法测试
# 王五.attack(赵三)
# 赵三.dead()
# 赵三.get_blood()

# 正式开始战斗
roles=[王五,金蛇郎君,赵三,李六,]

# 只要人没有只剩下一个或者零人，就必须持续战斗
while len(roles) not in (1,0):
    
    # 随机抽取2人
    role1,role2 = sample(roles,2)
    # print(role1.get_name(),role2.get_name())
    
    # 发动攻击方法
    role1.attack(role2)
    
    # 发起攻击后判断是否死亡 死亡就要移除
    roles = [role for role in roles if not role.dead()] # 列表生成式 调用方法判断是否死亡 没死的生成新列表继续战斗

if roles != []:
    print('最终的胜利者是：',roles[0].get_name())    
else:
    print('无人生还')
```

### 自定义类Class
###### 六、基于动态添加 继承等机制 实现对于python自有类的改造
（1）继承改造法 （2）动态添加 不建议对自有使用 c语言实现的不支持这么干哦 

```python
# 1.继承改造法
# 自定义类mylist
class Mylist(list):
    def print(self):
        for i in self:
            print(i)
            
# 展示自定义的方法，其他都是继承来的
l1 = Mylist([1,2,3,4])

# 自定义的使用
l1.print()

# 测试继承来的属性
print(l1)
```

```python
# 2. 动态添加法
# 定义一个函数
def new__str__(self):
    for i in self:
        return i
    
# 覆盖list原有的
list.__str__ = new__str__ # cannot set '__str__' attribute of immutable type 'list'


def myprint(self):
        for i in self:
            print(i)

# 给他增加到方法中
list.myprint = myprint # cannot set '__str__' attribute of immutable type 'list'
```

3.1 help 读取文档字符串**（__doc__）** 及其他类信息来获取信息解释一个函数/类          
只需要将定义类/函数的时候文档字符串只需要位于 类/函数的 定义语句class/def的下方用长字符串标识就可以被自动读取到当做__doc__    
##### 通常会包含以下信息：
1. 简要描述：在函数、类或模块的文档字符串的第一行应该是一个简短的描述语句，说明它是做什么的。
2. 参数和返回值：对于函数和方法，应该详细描述每个参数的类型和作用，以及函数的返回值。
3. 异常：如果函数可能抛出异常，应该在文档字符串中说明。
4. 用法示例：包含一个或多个简单的使用示例可以大大提高文档字符串的实用性。

```python
# 4. dir查看类的属性方法名称
# dir(Mylist)

# dir 是一个优先调用对象的__dir__方法的函数，没有就调用__dict__（无需显式定义）通常自动生成，除非指定 __slots__ 属性
class Mylist(list):
    def print(self):
        for i in self:
            print(i)
            
    def __dir__(self):
        return ['print']  # 定制的 dir 输出

dir(Mylist()) # 依旧全部输出，因为这个__dir__的定义只对实例生效 想要对于类生效 需要改元类，不建议
l1 = Mylist([1,2,3]) 
dir(l1) # ['print'] 
```

```python
# 定义文档字符串示例
def calculate_area(base, height, shape="triangle"):
    """
    Calculate the area of a geometric shape.

    This function supports the calculation of the area for triangles and rectangles.
    The default shape is 'triangle'.

    Parameters:
    - base (float): The base length of the geometric shape.
    - height (float): The height of the geometric shape.
    - shape (str, optional): The type of shape ('triangle' or 'rectangle'). Default is 'triangle'.

    Returns:
    float: The area of the specified geometric shape.

    Raises:
    ValueError: If the given shape is neither 'triangle' nor 'rectangle'.

    Examples:
    >>> calculate_area(10, 5)
    25.0
    >>> calculate_area(10, 5, shape="rectangle")
    50.0
    >>> calculate_area(10, 5, shape="circle")
    Traceback (most recent call last):
      ...
    ValueError: Shape must be either 'triangle' or 'rectangle'.
    """
    if shape == "triangle":
        return 0.5 * base * height
    elif shape == "rectangle":
        return base * height
    else:
        raise ValueError("Shape must be either 'triangle' or 'rectangle'.")
```
